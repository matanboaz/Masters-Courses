## Small Area Estimation - Home Assignment #08 Solution

### Q.1  

Write the AR(1) model in the general State-Space form (Equation (a) in the slides).  
Define the vectors $\beta_{ti},\,\eta_{ti}$, and the matrix $T_t$.

#### Solution  

The AR(1) model:  
$$\big(\beta_{tik}-\beta_{k}\big)=\phi_k\big(\beta{t-1,ik}-\beta_k\big)+\eta_{itk},\quad k=1,\ldots,K$$  

The State-Space model:  
$$\beta_{ti}=T_t\beta_{t-1,i}+\eta_{ti}$$  
where  

$$
\begin{equation*}
    \begin{pmatrix}
        \beta_{ti}-\beta \\
        \beta
    \end{pmatrix}= 
    \begin{pmatrix}
        \phi & 1-\phi \\
        0 & 1 
    \end{pmatrix}
    \begin{pmatrix}
        \beta_{t-1,i} \\
        \beta
    \end{pmatrix} + 
    \begin{pmatrix}
        \eta_{ti} \\
        0
    \end{pmatrix}
\end{equation*}
$$

### Q.2  

Write the Fay-Herriot model and the Battese _et al_. model in State-Space form.  
Define all the vectors and matrices.

#### Solution  

Fay-Herriot model as State-Space form:  
Notice that  
$$y_{ti}=\tilde{\theta}_{ti}=\theta_{ti}+e_{ti}=x_{ti}'\beta+u_{ti}+e_{ti}$$  
therefore,  
$$
\begin{equation*}
    y_{ti}=
    \begin{pmatrix}
        1 & x_{ti}'
    \end{pmatrix}
    \begin{pmatrix}
        u_{ti}  \\
        \beta
    \end{pmatrix}
    + e_{ti}
\end{equation*},\quad \sigma_e^2=\sigma^2
$$  

And we get:  

$$
\begin{equation*}
    \begin{pmatrix}
        u_{ti} \\
        \beta
    \end{pmatrix}= 
    \begin{pmatrix}
        0 & 0 \\
        0 & 1 
    \end{pmatrix}
    \begin{pmatrix}
        u_{t-1,i} \\
        \beta
    \end{pmatrix} + 
    \begin{pmatrix}
        u_{ti} \\
        0
    \end{pmatrix}
\end{equation*}
$$  

For Battese, Harter and Fuller:  

The original form is: $y_{ij}=x_{ij}'\beta+u_i+\varepsilon_{ij}$. Therefore,  
$$
\begin{align}
     \beta_{ti} & =  \begin{pmatrix}
        u_{ti} \\
        \beta
    \end{pmatrix} \\  
    Var\big(e_{ti}\big)  & = \sigma^2_\varepsilon  \\
    T_t  & =  \begin{pmatrix}
        0 & 0 \\
        0 & 1 
    \end{pmatrix}  \\
    \eta_{ti} & =  \begin{pmatrix}
        u_{i} \\
        0
    \end{pmatrix}
\label{eq:1}
\end{align}
$$

### Q.3  

Prove the expression $$Var\big(\hat{y}_{t+1|t}-y_{t+1}\big)=x_{t+1}'P_{t+1|t}x_{t+1}+\sigma^2_{t+1}$$  
for the general State-Space model (slide 16).

#### Solution  

$$
\begin{align}
    Var\big(\hat{y}_{t+1|t}-y_{t+1}\big) & = Var\big(x_{t+1}'\hat\beta_{t+1|t}-x_{t+1}'\beta_{t+1}-\varepsilon_{t+1}\big)  \\
        & = x_{t+1}'P_{t+1|t}x_{t+1}+Var(\varepsilon_{t+1}) \\
        & = x_{t+1}'P_{t+1|t}x_{t+1}+\sigma^2  \\
        & = x_{t+1}'P_{t+1|t}x_{t+1}+\sigma^2_{t+1}
\end{align}
$$

### Q.4  

Consider the Fay-Herriot model $Y_t=\alpha_t+z_t\gamma+\varepsilon_t$ where the index $t$ denotes an area, $z_t$ is a known scalar and $\alpha_t$ and $\varepsilon_t$ are independent white noise series (i.i.d.), with zero mean and variance $Var(\varepsilon_t)=\sigma_\varepsilon^2,\,Var(\alpha_t)=\sigma_\alpha^2$. (Model with random intercept).  

#### (a)  

Write the model in State-Space form.  

#### Solution  

$$
\begin{equation*}
    Y_t = 
        \begin{pmatrix}
            1  &  z_t
        \end{pmatrix}
        \begin{pmatrix}
            \alpha_t  \\
            \gamma
        \end{pmatrix}
        + \varepsilon_t,\quad Var(\varepsilon_t)=\sigma^2_{\varepsilon}
\end{equation*}
$$
and for $\beta_t=T_t\beta_{t-1}+\eta_t$:  
$$
\begin{equation*}
    \begin{pmatrix}
        \alpha_t \\
        \gamma
    \end{pmatrix} = 
    \begin{pmatrix}
        0  &  0  \\
        0  &  1
    \end{pmatrix}
    \begin{pmatrix}
        \alpha_{t-1}  \\
        \gamma
    \end{pmatrix} +  
    \begin{pmatrix}
        \alpha_t  \\
        0
    \end{pmatrix},\quad Q=\begin{pmatrix}
                              \sigma^2_\alpha  &  0  \\
                              0                &  0
                          \end{pmatrix}
\end{equation*}
$$

#### (b)  

Suppose that $z_{12}=20$ and that at time $11$, $\hat\alpha_{11}=4.2$ and $\hat\gamma_{(11)}=0.5$. (The estimators of $\alpha$ and $\gamma$ at time 11).  
Show that the predictor of $\alpha_{12}$ at time $t=11$ is $\hat\alpha_{12|11}=0$ and predict $\hat Y_{12|11}$.

#### Solution   

$$
\begin{align}
    \hat\beta_{12|11}  &  =  T_{12}\hat\beta_{11}  \\  
                       &  =  \begin{pmatrix}
                                 0  &  0  \\
                                 0  &  1
                             \end{pmatrix}
                             \begin{pmatrix}
                                 4.2 \\
                                 0.5
                             \end{pmatrix}  \\
                        &  = \begin{pmatrix}
                                 0  \\
                                 0.5
                             \end{pmatrix} \\  
                        &  = \begin{pmatrix}
                                 \hat\alpha_{12|11}  \\
                                 \hat\gamma_{12|11}
                             \end{pmatrix}
\end{align}
$$  

Therefore, $$
\hat\alpha_{12|11} = 0
$$
Hence, $$
\hat Y_{12|11}=x_{12}'\hat\beta_{12|11}=
\begin{pmatrix}
    1  &  20
\end{pmatrix}
\begin{pmatrix}
    0  \\
    0.5
\end{pmatrix} = 10
$$

#### (c)  

Compute the variance of the corresponding prediction error if $\sigma_\varepsilon^2=0.4$, $\sigma_\alpha^2=0.07$ and 
$$
\begin{equation*}
    P_{11} =
        \begin{pmatrix}
            \hat\alpha_{11}-\alpha_{11}  \\
            \hat\gamma_{(11)}-\gamma
        \end{pmatrix}
        \begin{pmatrix}
            \big(\hat\alpha_{11}-\alpha_{11}\big)  &  \big(\hat\gamma_{(11)}-\gamma\big)
        \end{pmatrix} =
        \begin{pmatrix}
            0.05  &  0.02  \\
            0.02  & 0
        \end{pmatrix}

$$

#### Solution   


Notice that  $$P_{12|11}=T_{12}P_{11}T^T_{12}+Q_{12}$$
Therefore,
$$
\begin{align}
    P_{12|11}  &  =  
        \begin{pmatrix}
            0  &  0  \\
            0  &  1
        \end{pmatrix}
        \begin{pmatrix}
            0.05  &  0.02  \\
            0.02  &  0.1
        \end{pmatrix}
        \begin{pmatrix}
            0  &  0  \\
            0  &  1
        \end{pmatrix}  +  
        \begin{pmatrix}
            0.07  &  0  \\
            0  &  1
        \end{pmatrix}  \\
      &  =  
      \begin{pmatrix}
          0  &  0  \\
          0  &  0.1
      \end{pmatrix}  +
      \begin{pmatrix}
          0.07  &  0  \\
          0  &  0
      \end{pmatrix}  \\
      &  =  \begin{pmatrix}
                0.07  &  0  \\
                0  &  0.1
            \end{pmatrix}
\end{align}
$$  

Now,  

$$
\begin{align}
    Var\big(\hat Y_{12|11}-Y_{12}\big)  &  =  x_{12}'P_{12|11}x_{12} + \sigma_{12}^2 = f_{12}  \\
        &  =
        \begin{pmatrix}
            1  &  20
        \end{pmatrix}
        \begin{pmatrix}
            0.07  &  0  \\
            0  &  0.1
        \end{pmatrix}
        \begin{pmatrix}
            1  \\
            20
        \end{pmatrix} + 0.4  \\
        &  =  40.07 + 0.4  \\
        &  =  40.47
\end{align}
$$

## Q.5  

Write the Rao and Yu model (Slide 25) for a single area as a State-Space model for the case where the sampling errors are independent over time.

#### Solution  


First notice that the Rao and Yu model is 
$$
\begin{align}
    y_{ti}  &  =  \theta_{ti} + e_{ti}  \\
            &  =  x_{ti}'\beta + u_i + v_{ti} + e_{ti}
\end{align}
$$
along with  $$v_{ti} = \rho v_{t-1,i} + \zeta_{ti}$$  

Now,  

$$
\begin{equation*}
    y_{ti} = 
        \begin{pmatrix}
            1  &  1  &  x_{ti}'
        \end{pmatrix}
        \begin{pmatrix}
            u_i  \\
            v_{ti}  \\
            \beta
        \end{pmatrix} + e_{ti}
\end{equation*}
$$

$$
\begin{equation*}
    \beta_t=T_t\beta_{t-1}+\eta_{ti}
\end{equation*}
$$

Therefore,  

$$
\begin{equation*}
    \begin{pmatrix}
        u_i \\
        v_{ti}  \\
        \beta
    \end{pmatrix} = 
        \begin{pmatrix}
            1  &  0  &  0  \\
            0  &  \rho  &  0  \\
            0  &  0  &  1
        \end{pmatrix}
        \begin{pmatrix}
            u_i  \\
            v_{t-1,i}  \\
            \beta
        \end{pmatrix}  
        +  \begin{pmatrix}
               0  \\
               \zeta_{ti}  \\
               0
           \end{pmatrix}
\end{equation*}
$$  


$$
\begin{equation*}
    Q = \begin{pmatrix}
            0  \\
            \zeta_{ti}  \\
            0
        \end{pmatrix}
        \begin{pmatrix}
            0  &  \zeta_{ti}  &  0
        \end{pmatrix}
      = \begin{pmatrix}
            0  &  0  &  0  \\
            0  &  \sigma^2_\zeta  &  0  \\
            0  &  0  &  0
        \end{pmatrix}
\end{equation*}
$$
